```python
from bs4 import BeautifulSoup
import urllib.request
list_link =[]
def getPage(i):
  url =  f'https://vietnamnet.vn/thoi-su/chinh-tri-page{i}'
  page = urllib.request.urlopen(url)
  soup = BeautifulSoup(page, 'html.parser')
  news = soup.findAll('div', class_='feature-box__content')
  for new in news:
    title = new.find('h3',class_= 'feature-box__content--title vnn-title').find('a')
    link = title.get('href')
    list_link.append(link)
```

```python
for i in range(1,200,10):
    getPage(i)
```

```python
list_link
```

```python
links=[]
```

```python
for i in range(30,len(list_link)):
    links.append(list_link[i])
```

```python
Title = []
Content=[]
def get_news(url):
  page = urllib.request.urlopen(url) 
  soup = BeautifulSoup(page,'html.parser')
  try:
    tit = soup.find('h1',class_='newsFeature__header-title mt-20').getText()
  except:
    tit = None
  try:
   content = soup.find('div',class_='maincontent').getText()
  except:
    content =None
  Title.append(tit)
  Content.append(content)
```

```python
for i in range(len(links)):
  try:
   get_news(links[i])
  except:
    print('link has error',i)
```

```python
def Remove_n(r):
    r=  r.replace("\r\n", "")
    r = r.replace("\n\r", "")
    r = r.replace("\n\n", "")
    r = r.replace("\n\n", "")
    r = r.replace("\n", "")
    r = r.replace(".\n", ". ")
    return r
```
