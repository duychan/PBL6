```python
list_link=[]
for i in range(1,170,5):  
    url =  f'https://thanhnien.vn/thoi-su/chinh-tri/?trang={i}'
    page = urllib.request.urlopen(url)
    soup = BeautifulSoup(page, 'html.parser').find('div',class_="relative")
    news = soup.findAll('article', class_='story')
    for new in news:
        title = new.find('a',class_='story__thumb')
        link=title.get('href')
        list_link.append(link)
```

```python
list_link
```

```python
Title = []
Content=[]
def get_news(url):
  page = urllib.request.urlopen(url) 
  soup = BeautifulSoup(page,'html.parser')
  try:
    tit = soup.find('h1',class_='details__headline cms-title').getText()
  except:
    tit = None
  try:
   content = soup.find('div',class_='details__content').getText()
  except:
    content =None
  Title.append(tit)
  Content.append((content))
```

```python
import time
```

```python
for link in list_link:
    get_news(link)
    time.sleep(2)
```

```python
def Remove_n(r):
    r=  r.replace("\r\n", "")
    r = r.replace("\n\r", "")
    r = r.replace("\n\n", "")
    r = r.replace("\n\n", "")
    r = r.replace("\n", "")
    r = r.replace(".\n", ". ")
    return r
```
