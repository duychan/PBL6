```python
from bs4 import BeautifulSoup
import urllib.request
```

```python
url =  f'https://dantri.com.vn/xa-hoi/chinh-tri/trang-{3}.htm'
page = urllib.request.urlopen(url)
```

```python
soup = BeautifulSoup(page, 'html.parser')
```

```python
news = soup.findAll('div', class_='article-thumb')
```

```python
list_link =[]
for new in news:   
    title = new.find('a')
    link = title.get('href')
    list_link.append(link)
```

```python

list_link =[]
def getPage(i):
  url =  f'https://dantri.com.vn/xa-hoi/chinh-tri/trang-{3}.htm'
  page = urllib.request.urlopen(url)
  soup = BeautifulSoup(page, 'html.parser')
  news = soup.findAll('div', class_='article-thumb')
  for new in news:
    title = new.find('a')
    link = title.get('href')
    list_link.append(link)
```

```python
for i in range(1,200,5):
    getPage(i)
```

```python
Title = []
Content=[]
def get_news(url):
  page = urllib.request.urlopen('https://dantri.com.vn/'+url) 
  soup = BeautifulSoup(page,'html.parser')
  try:
    tit = soup.find('h1',class_='title-page detail').getText()
  except:
    tit = None
  try:
   content = soup.find('div',class_='singular-content').getText()
  except:
    content =None
  Title.append(tit)
  Content.append((content))
```

```python
for link in list_link:
    get_news(link)
```

```python
Title
```

```python
links=[]
for link in list_link:
    links.append('https://dantri.com.vn'+link)
```
