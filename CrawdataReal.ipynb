```python
Title=[]
Content=[]
for urli in all_link:
  page = driver.get(urli)
  soup = BeautifulSoup(driver.page_source,'html.parser')
  try:
    tit = soup.find('h1',class_='article__title cms-title').getText()
  except:
    tit = None
  try:
   content = soup.find('div',class_='article__body cms-body').getText()
  except:
    content =None
  Title.append(tit)
  Content.append(content)
  time.sleep(3)
```
